# 1.0 The context and understanding the business

## 1.1 O que é Fraude?

Fraude é uma ação intencionalmente enganosa projetada para **proporcionar ao perpetrador um ganho ilegal ou negar um direito à vítima**. Além disso, é **um ato deliberado (ou falha em agir) com a intenção de obter um benefício não autorizado**, seja para si próprio ou para a instituição, utilizando engano ou falsas sugestões ou supressão da verdade ou outros meios antiéticos, que são acreditados e confiados por outros. Privar outra pessoa ou a instituição de um benefício ao qual ele/ela/ele tem direito, usando qualquer um dos meios descritos acima, também constitui fraude.

Tipos de fraude incluem fraude fiscal, fraude de cartão de crédito, fraude eletrônica, fraude de valores mobiliários e fraude de falência. Atividades fraudulentas podem ser realizadas por um indivíduo, vários indivíduos ou uma empresa como um todo.

**Exemplos de atos fraudulentos incluem**, mas não se limitam a, o seguinte:

- Desfalque.
- Falsificação ou alteração de documentos.
- Alteração ou manipulação não autorizada de arquivos de computador.
- Relatórios financeiros fraudulentos.
- Apropriação indevida ou uso indevido de recursos da Universidade (por exemplo, fundos, suprimentos, equipamentos, instalações, serviços, inventário ou outros ativos).
- Autorização ou recebimento de pagamento por bens não recebidos ou serviços não realizados.
- Autorização ou recebimento de salários ou benefícios não ganhos.
- Conflito de interesse, violações de ética.

**Fraude envolve a falsa representação de fatos**, seja por intencionalmente omitir informações importantes ou fornecer declarações falsas a outra parte com o objetivo específico de obter algo que não teria sido fornecido sem o engano.

Freqüentemente, o perpetrador de fraude está ciente de informações que a vítima pretendida não está, permitindo que o perpetrador engane a vítima. No fundo, o indivíduo ou empresa que comete fraude está se aproveitando da assimetria de informações; especificamente, que o custo de recurso de revisar e verificar essas informações pode ser significativo o suficiente para criar um desincentivo a investir totalmente na prevenção de fraudes.

Tanto os estados quanto o governo federal possuem leis que criminalizam a fraude, embora ações fraudulentas nem sempre resultem em um julgamento criminal. Promotores governamentais geralmente têm uma discricionariedade substancial em determinar se um caso deve ir a julgamento e podem buscar um acordo em vez disso, se isso resultar em uma resolução mais rápida e menos custosa. Se um caso de fraude for a julgamento, o perpetrador pode ser condenado e enviado para a prisão.


## 1.2 Tipos de transações

Uma **transação financeira** é um acordo ou comunicação, realizado entre um comprador e um vendedor para trocar um ativo por pagamento.

Envolve uma mudança no status das finanças de duas ou mais empresas ou indivíduos. O comprador e o vendedor são entidades ou objetos separados, muitas vezes envolvendo a troca de itens de valor, como informações, bens, serviços e dinheiro. Ainda é uma transação se os bens forem trocados em um momento e o dinheiro em outro. Isso é conhecido como uma transação em duas partes: a primeira parte é entregar o dinheiro, a segunda parte é receber os bens.


### 1.2.1 Transações em Dinheiro - Entrada e Saída de Dinheiro

Uma transação em dinheiro refere-se a uma transação que envolve **uma saída imediata de dinheiro para a compra** de quaisquer bens, serviços ou ativos. Transações em dinheiro podem ser orientadas ao consumidor ou orientadas a negócios.

Uma transação em dinheiro se diferencia de outros modos de pagamento, como transações de crédito em um negócio que envolve contas a receber. Da mesma forma, uma transação em dinheiro também é diferente de transações com cartão de crédito.

Transações em dinheiro são diferentes de transações que envolvem um atraso na entrega dos bens ou um atraso no pagamento. Tais transações incluem venda a crédito, contrato a termo, contrato futuro e outras transações com margem.


### 1.2.2 Débito


Um pagamento com cartão de débito é o mesmo que um pagamento imediato em dinheiro, pois o valor é **deduzido instantaneamente da sua conta bancária**.

Cartões de débito permitem que os clientes bancários gastem dinheiro retirando fundos existentes que já depositaram no banco, como de uma conta corrente. Uma transação de débito usando seu PIN (número de identificação pessoal), \textbf{é uma transação online concluída em tempo real}. Quando você completa uma transação de débito, você autoriza a compra com seu PIN e o comerciante comunica imediatamente com seu banco ou cooperativa de crédito, fazendo com que os fundos sejam transferidos em tempo real.

O primeiro cartão de débito pode ter entrado no mercado já em 1966, quando o Bank of Delaware testou a ideia.


### 1.2.3 Pagamento

Um ato iniciado pelo pagador ou pelo beneficiário, ou em nome do pagador, de colocar, transferir ou retirar fundos, independentemente de quaisquer obrigações subjacentes entre o pagador e o beneficiário.


### 1.2.4 Transferência

Uma transferência envolve o **movimento de ativos, fundos monetários e/ou direitos de propriedade de uma conta para outra**. Uma transferência pode exigir uma troca de fundos quando envolve uma mudança de propriedade, como quando um investidor vende um imóvel. Nesse caso, há uma transferência de título do vendedor para o comprador e uma transferência simultânea de fundos, igual ao preço negociado, do comprador para o vendedor.

O termo transferência também pode se referir ao movimento de uma conta de um banco ou corretora para outra.


## 1.3 Fatores chave

- Fraude envolve **engano** com a intenção de ganhar ilegal ou antiético às custas de outro.
- Em **finanças**, a fraude pode assumir muitas formas, incluindo a realização de falsos pedidos de seguro, manipulação de dados contábeis, esquemas de pump & dump e roubo de identidade levando a compras não autorizadas.
- A fraude **custa bilhões de dólares** à economia a cada ano, e aqueles que são pegos estão sujeitos a multas e tempo de prisão.
- A **fraude ao consumidor** ocorre quando uma pessoa sofre uma perda financeira envolvendo o uso de práticas comerciais enganosas, injustas ou falsas.
- Com o **roubo de identidade**, os ladrões roubam suas informações pessoais, assumem sua identidade, abrem cartões de crédito, contas bancárias e fazem compras.
- As **fraudes hipotecárias** visam proprietários em dificuldades para obter dinheiro deles.
- A fraude com **cartão de crédito e débito** ocorre quando alguém obtém suas informações do cartão e faz compras ou oferece para reduzir a taxa de juros do seu cartão de crédito.
- **Falsas instituições de caridade** e loterias se aproveitam da simpatia ou ganância das pessoas.
- A fraude de **cobrança de dívidas** tenta cobrar contas não pagas, sejam elas suas ou não.
- As **fraudes relacionadas à COVID-19** são um novo tipo de fraude projetado para se aproveitar do seu medo ou necessidade financeira.


## 1.4 Considerações legais


Enquanto o governo pode decidir que um caso de fraude pode ser resolvido fora dos processos criminais, partes não governamentais que alegam prejuízo podem buscar uma ação civil. As vítimas de fraude podem processar o perpetrador para recuperar fundos ou, em um caso em que não houve perda monetária, podem processar para restabelecer os direitos da vítima.

Provar que a fraude ocorreu requer que o perpetrador tenha cometido atos específicos. Primeiro, o perpetrador tem que fornecer uma declaração falsa como um fato material. Segundo, o perpetrador teve que saber que a declaração era falsa. Terceiro, o perpetrador teve a intenção de enganar a vítima. Quarto, a vítima tem que demonstrar que confiou na declaração falsa. E quinto, a vítima teve que ter sofrido danos como resultado de agir com base na declaração intencionalmente falsa.


## 1.5 Consequências da Fraude Financeira


Primeiro, servindo como um **sinal de desonestidade**, a fraude financeira faz com que clientes e fornecedores duvidem dos compromissos de uma empresa no mercado de produtos, enfraquecendo os incentivos para que clientes e fornecedores assinem contratos com a empresa. Em segundo lugar, a fraude financeira afeta diretamente **as capacidades de financiamento de uma empresa e os custos de financiamento**, bem como os ajustes da governança corporativa (como a saída de executivos). Isso leva a grandes dificuldades e incertezas nas atividades de produção e operação de uma empresa. Assim, é impossível para empresas fraudulentas cumprir seus compromissos existentes (ou futuros compromissos) (Cornell & Shapiro, 1987).

De acordo com a revista Infosecurity, **a fraude custou à economia global £3,2 trilhões em 2018**. Para algumas empresas, as perdas com fraudes ultrapassam 10% de seus gastos totais. Essas perdas massivas levam as empresas a buscar novas soluções para prevenir, detectar e eliminar fraudes.

![](img/fraud_losses.png)

A fraude pode ter um impacto devastador em um negócio. Em 2001, foi descoberta uma **fraude corporativa massiva** na Enron, uma empresa de energia sediada nos Estados Unidos. Executivos usaram uma variedade de técnicas para disfarçar a saúde financeira da empresa, incluindo a obstrução deliberada de receitas e a representação incorreta de lucros. Após a descoberta da fraude, os acionistas viram os preços das ações despencarem de cerca de $90 para menos de $1 em pouco mais de um ano. Os funcionários da empresa tiveram seu patrimônio eliminado e perderam seus empregos após a Enron declarar falência. O escândalo da Enron foi um dos principais motivadores por trás das regulamentações encontradas na **Lei Sarbanes-Oxley** aprovada em 2002.

Comparadas com as empresas de controle, as empresas envolvidas em fraudes financeiras apresentam uma **queda na receita de vendas** de 11,9–17,1% e uma **diminuição em sua margem de lucro bruto** nas vendas de 2,4–2,8% nos três anos após a punição. Além disso, a receita de vendas dos cinco principais clientes grandes cai 43,9–55,1% no período pós-punição, enquanto a receita de vendas dos clientes pequenos não diminui significativamente.


## 1.6 A Empresa de Bloqueio de Fraudes - FBC


A Empresa de Bloqueio de Fraudes (FBC) é uma empresa especializada em detectar fraudes em transações financeiras feitas por meio de dispositivos móveis. A empresa possui um serviço chamado "Fraude Blocker", no qual garante o bloqueio de transações fraudulentas.

O modelo de negócios da empresa é do tipo Serviço, com a monetização feita pelo desempenho do serviço prestado, ou seja, o cliente paga uma taxa fixa pelo sucesso na detecção de fraudes nas transações do cliente. Além disso, a FBC está expandindo e, para adquirir clientes mais rapidamente, adotou uma estratégia muito agressiva. A FBC irá:

- Receber 25% do valor de cada transação que for realmente detectada como fraude (verdadeiros positivos).
- Receber 5% do valor de cada transação detectada como fraude, mas a transação é realmente legítima (falsos positivos).
- Reembolsar 100% do valor ao cliente, para cada transação detectada como legítima, no entanto a transação é realmente uma fraude (falsos negativos).

Com essa estratégia agressiva, a empresa assume os riscos de não detectar fraudes e é remunerada pela detecção assertiva de fraudes. Para o cliente, é um excelente negócio contratar a FBC. Embora a taxa cobrada seja muito alta em caso de sucesso, 25%, a empresa reduz seus custos com transações fraudulentas corretamente detectadas e os danos causados por um erro no serviço antifraude serão cobertos pela própria FBC.

Para a FBC, além de conseguir muitos clientes com essa estratégia arriscada para garantir o reembolso em caso de falha na detecção de fraude do cliente, depende apenas da precisão e acurácia dos modelos construídos por seus Cientistas de Dados. Ou seja, quanto maior a precisão do modelo "Fraud Blocker", maior será a receita da empresa. No entanto, se o modelo tiver baixa precisão, a empresa poderá ter uma grande perda.

### 4.4.3 Statistical relationships

As we observed in the previous section, there are outliers in the data set. We're going to check which values can be considered as outliers by calculating the Q1, Q3 and IQR.

```python
# calculates Q1, Q3 and IQR
Q1 = np.log1p(np.quantile(is_fraud['amount'], .25))
Q3 = np.log1p(np.quantile(is_fraud['amount'], .75))
IQR = Q3 - Q1

# calculates the outliers boundaries through statistical relationship
low_outliers = Q1 - 1.5 * IQR
high_outliers = Q3 + 1.5 * IQR

print("Fraud amount boundaries for outliers")
print(f"\nQ1 = {Q1}\nQ3 = {Q3}\nIQR = {IQR}\n\nlow_outliers = {low_outliers}\nhigh_outliers = {high_outliers}")
```

As observed, low outliers identified by statistical relationships are close to the values that we identified in the previous plots (anything lower than `6 ($402.43)`).

# 5.0 Feature Engineering

## 5.1 Loading data set

```python
# defines the new schema: field, type, nullabe or not
data_schema = [StructField('step', ShortType(), True), 
               StructField('type', StringType(), True),
               StructField('amount', DoubleType(), True),
               StructField('name_orig', StringType(), True),
               StructField('old_balance_orig', DoubleType(), True),
               StructField('new_balance_orig', DoubleType(), True),
               StructField('name_dest', StringType(), True),
               StructField('old_balance_dest', DoubleType(), True),
               StructField('new_balance_dest', DoubleType(), True),
               StructField('is_fraud', ShortType(), True),
               StructField('is_flagged_fraud', ShortType(), True)]

# defines the final struct
final_struct = StructType(fields=data_schema)

# loads data set in spark
df_fraud_spark = spark.read.csv('data/fraud_clean.csv', schema=final_struct, header=True)
```

Features to create:
- `orig_type` (customer or merchant)
- `dest_type` (customer or merchant)
- `old_balance_orig_status` (zero or non zero)
- `new_balance_orig_status` (zero or non zero)
- `old_balance_dest_status` (zero or non zero)
- `new_balance_dest_status` (zero or non zero)
- `is_orig_old_balance_greater_new` (yes or no)
- `is_dest_old_balance_greater_new` (yes or no)
- `transaction_direction` (C2C, C2M, M2C, M2M)
- `is_orig_equal_dest` (yes or no)

## 5.2 Origin Type and Destination Type

As we have the transaction origin that is identfied by starting with 'C' (Customer) or 'M' (Merchant) we can categorize the transaction origin.

```python
# creates the new feature: orig_type
df_fraud_spark = df_fraud_spark.withColumn("orig_type",
                                           when(df_fraud_spark.name_orig.like('C%'), "customer")
                                           .otherwise("merchant"))

# creates the new feature: dest_type
df_fraud_spark = df_fraud_spark.withColumn("dest_type", 
                                           when(df_fraud_spark.name_dest.like('C%'), "customer")
                                           .otherwise("merchant"))
```

## 5.3 Origin Old balance and New balance  status

Now let's create a feature that will tell us if the destination old and new balance is zero or non-zero.

```python
# creates the new feature: old_balance_orig_status
df_fraud_spark = df_fraud_spark.withColumn("old_balance_orig_status", 
                                           when(col("old_balance_orig") > 0, "non_zero")
                                           .otherwise("zero"))

# creates the new feature: new_balance_orig_status
df_fraud_spark = df_fraud_spark.withColumn("new_balance_orig_status", 
                                           when(col("new_balance_orig") > 0, "non_zero")
                                           .otherwise("zero"))
```

## 5.4 Creating Destination Old balance and New balance status

Now let's create a feature that will tell us if the destination old and new balance is zero or non-zero.

```python
# creates the new feature: old_balance_dest_status
df_fraud_spark = df_fraud_spark.withColumn("old_balance_dest_status", 
                                           when(col("old_balance_dest") > 0, "non_zero")
                                           .otherwise("zero"))

# creates the new feature: new_balance_dest_status
df_fraud_spark = df_fraud_spark.withColumn("new_balance_dest_status", 
                                           when(col("new_balance_dest") > 0, "non_zero")
                                           .otherwise("zero"))
```

## 5.5 Is origin equal to destination?

Now let's create a feature that will tell us if the origin is equal to the destination.

```python
# creates the new feature: is_orig_equal_dest
df_fraud_spark = df_fraud_spark.withColumn("is_orig_equal_dest", 
                                           when(col("name_orig") == col("name_dest"), "yes")
                                           .otherwise("no"))
```

## 5.6 Is the origin old balance greather than the new?

Now let's create a feature that will tell us if the old balance from the origin is greater than the new.

```python
# creates the new feature: is_orig_old_balance_greater_new
df_fraud_spark = df_fraud_spark.withColumn("is_orig_old_balance_greater_new", 
                                           when(col("old_balance_orig") > col("new_balance_orig"), "yes")
                                           .otherwise("no"))
```

## 5.7 Is the destinarion old balance greather than the new?

Now let's create a feature that will tell us if the old balance from the destination is greater than the new.

```python
# creates the new feature: is_dest_old_balance_greater_new
df_fraud_spark = df_fraud_spark.withColumn("is_dest_old_balance_greater_new", 
                                           when(col("old_balance_dest") > col("new_balance_dest"), "yes")
                                           .otherwise("no"))
```

## 5.8 Transaction direction

As we have different origin and destination types, let's create a feature that will tell us:

- C2C, if the transaction is Customer to Customer
- C2M, if the transaction is Customer to Merchant
- M2C, if the transaction is Merchant to Customer
- M2M, if the transaction is Merchant to Merchant

```python
# creates the new feature: transaction_direction
df_fraud_spark = df_fraud_spark.withColumn("transaction_direction", 
                                           when((df_fraud_spark.name_orig.like('C%') & df_fraud_spark.name_dest.like('C%')), "C2C")
                                           .when((df_fraud_spark.name_orig.like('C%') & df_fraud_spark.name_dest.like('M%')), "C2M")
                                           .when((df_fraud_spark.name_orig.like('M%') & df_fraud_spark.name_dest.like('C%')), "M2C")
                                           .otherwise("M2M"))
```

## 5.9 Changing `is_flagged_fraud` to categorical

```python
# changes is_flagged_fraud to categorical
df_fraud_spark = df_fraud_spark.withColumn("is_flagged_fraud", 
                                           when(col("is_flagged_fraud") == 1, "yes")
                                           .otherwise("no"))
```

- Keep the classes proportions
- Get a statistically significant sample size

- **Population size:** 6,362,620 (that's the full data set size)
- **Confidence level:** 95% (that's industry standard)
- **Margin of error:** 0.1%

More info at: https://www.surveymonkey.com/mp/sample-size-calculator/

```python
# gets the number of fraud cases
sample_is_fraud = sample['is_fraud'].value_counts()[1]

# gets the number of non fraud cases
sample_is_not_fraud = sample['is_fraud'].value_counts()[0]
```

As observed, both algorithms have nearly the same behavior. We need to take into account that the test set presents an extremely imbalance, that's why the density for `0` it much much higher than for `1`.

Other metric that can help us assess the right algorithm for our project is the **Brier Score Loss**, according to its documentation:

> "The smaller the Brier score, the better, hence the naming with “loss”. Across all items in a set N predictions, the Brier score measures the mean squared difference between (1) the predicted probability assigned to the possible outcomes for item i, and (2) the actual outcome. Therefore, the lower the Brier score is for a set of predictions, the better the predictions are calibrated. Note that the Brier score always takes on a value between zero and one, since this is the largest possible difference between a predicted probability (which must be between zero and one) and the actual outcome (which can take on values of only 0 and 1). The Brier loss is composed of refinement loss and calibration loss."

Reference: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html

```python
# calculater the Brier Score Loss
bsl_xgb = brier_score_loss(y_test, y_pred_xgb_prob, pos_label=1)
bsl_cat = brier_score_loss(y_test, y_pred_cat_prob, pos_label=1)

# prints the calculated BSL for each algorithm probability
print(f'Brier Score Loss (XGBoost): {bsl_xgb}')
print(f'Brier Score Loss (CatBoost): {bsl_cat}')
```

As observed, the Brier Score Loss are different. For this score, the closer to `0`, the better. Thus, `XGB` suits best for our needs.

# 14.0 Business performance¶

```python
# sets the transaction numbers
num_total_transactions = 6362620
num_total_fraud_transactions = 8097
num_total_non_fraud_transactions = num_total_transactions - num_total_fraud_transactions

print(f'Total number of transactions: {num_total_transactions:,}')
print(f'Total number of fraud transactions: {num_total_fraud_transactions:,}')
print(f'Total number of non fraud transactions: {num_total_non_fraud_transactions:,}')


# sets the metrics
avg_model_recall = 0.9570
model_std = 0.0114
upper_model_recall = avg_model_recall + model_std  
lower_model_recall = avg_model_recall - model_std


# from the total actual fraud transactions, how many the model got right? (TP)
upper_num_TP_transactions = num_total_fraud_transactions * upper_model_recall

# from the total number of actual non fraud transactions, how many the model said it was fraud? (FP)
upper_num_FP_transactions = num_total_non_fraud_transactions * (1 - upper_model_recall)

# from the total number of actual fraud transactions, how many the model sait it was not fraud? (FN)
upper_num_FN_transactions = num_total_fraud_transactions * (1 - upper_model_recall)

print(f'\nBest scenario:')
print(f'From the total actual fraud transactions, how many the model got right? (TP) -> {round(upper_num_TP_transactions):,}')
print(f'From the total number of actual non fraud transactions, how many the model said it was fraud? (FP) -> {round(upper_num_FP_transactions):,}')
print(f'From the total number of actual non fraud transactions, how many the model said it was fraud? (FN) -> {round(upper_num_FN_transactions):,}')


# from the total actual fraud transactions, how many the model got right? (TP)
lower_num_TP_transactions = num_total_fraud_transactions * lower_model_recall

# from the total number of actual non fraud transactions, how many the model said it was fraud? (FP)
lower_num_FP_transactions = num_total_non_fraud_transactions * (1 - lower_model_recall)

# from the total number of actual fraud transactions, how many the model sait it was not fraud? (FN)
lower_num_FN_transactions = num_total_fraud_transactions * (1 - lower_model_recall)

print(f'\nWorst scenario:')
print(f'From the total actual fraud transactions, how many the model got right? (TP) -> {round(lower_num_TP_transactions):,}')
print(f'From the total number of actual non fraud transactions, how many the model said it was fraud? (FP) -> {round(lower_num_FP_transactions):,}')
print(f'From the total number of actual non fraud transactions, how many the model said it was fraud? (FN) -> {round(lower_num_FN_transactions):,}')
```

Let's recall the FBC business model.

- Receive 25% of the value of each transaction that is truly detected as fraud (true positives).
- Receive 5% of the value of each transaction detected as fraud, but the transaction is truly legitimate (false positives).
- Refund 100% of the value to the customer, for each transaction detected as legitimate, however the transaction is truly a fraud (false negatives).

Considering the parameters:
- **Median amount of a transaction:** \$7,4871.8 (we're using the median because the amount distribution is highly skewed)
- **Portfolio:** 6,362,620 transactions (fraudulent + non fraudulent)

#### True positive amount = [0.25 x (median amount) x (number of TP transactions)]
#### False positive amount = [0.05 x (median amount) x (number of FP transactions)]
#### False negative amount = [1 x (median amount) x (number of FN transactions)]

### Total revenue =  True positive amount + False positive amount - False negative amount

```python
# median transaction amount
median_transaction_amount = 74871.8

print(f'Median transaction amount: ${median_transaction_amount:,.2f}')


upper_TP_amount = 0.25 * median_transaction_amount * upper_num_TP_transactions
upper_FP_amount = 0.05 * median_transaction_amount * upper_num_FP_transactions
upper_FN_amount = 1 * median_transaction_amount * upper_num_FN_transactions

upper_total_revenue = upper_TP_amount + upper_FP_amount - upper_FN_amount

print(f'\nBest scenario total revenue = ${upper_total_revenue:,.2f}')


lower_TP_amount = 0.25 * median_transaction_amount * lower_num_TP_transactions
lower_FP_amount = 0.05 * median_transaction_amount * lower_num_FP_transactions
lower_FN_amount = 1 * median_transaction_amount * lower_num_FN_transactions

lower_total_revenue = lower_TP_amount + lower_FP_amount - lower_FN_amount

print(f'\nWorst scenario total revenue = ${lower_total_revenue:,.2f}')
```
